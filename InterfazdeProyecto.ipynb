```python
import cv2
import numpy as np

keyboard = np.zeros((600,1000, 3), np.uint8)

keys_set_1 = {0: "Q", 1: "W", 2: "E", 3: "R", 4: "T",
              5: "A", 6: "S", 7: "D", 8: "F", 9: "G",
              10: "Z", 11: "X", 12: "C", 13: "V", 14: "B"}

def letras(index , text, state):
    if index == 0:
        x = 0
        y = 0
    elif index == 1:
        x = 200
        y = 0
    elif index == 2:
        x = 400
        y = 0
    elif index == 3:
        x = 600
        y = 0
    elif index == 4:
        x = 800
        y = 0
    elif index == 5:
        x = 0
        y = 200
    elif index == 6:
        x = 200
        y = 200
    elif index == 7:
        x = 400
        y = 200
    elif index == 8:
        x = 600
        y = 200
    elif index == 9:
        x = 800
        y = 200
    elif index == 10:
        x = 0
        y = 400
    elif index == 11:
        x = 200
        y = 400
    elif index == 12:
        x = 400
        y = 400
    elif index == 13:
        x = 600
        y = 400
    elif index == 14:
        x = 800
        y = 400
    
    
    width = 200
    height = 200
    th = 3  #grosor de la linea

    if state is True:
    # cv2.rectangle(keyboard, posicion, tamano, color, grosor)
        cv2.rectangle(keyboard, (x+th, y+th), (x+width-th, y+height-th), (255,255,255), -1)  # dibujamos el rectangulo
    else: 
        cv2.rectangle(keyboard, (x+th, y+th), (x+width-th, y+height-th), (255,0,0), th )  # dibujamos el rectangulo

    
    
    # dibujamos la letra y lo acomodamos al centro del rectangulo 
    font_letter = cv2.FONT_HERSHEY_PLAIN
    font_scale = 10
    font_th = 4
    text_size = cv2.getTextSize(text, font_letter, font_scale, font_th)[0]
    width_text, height_text = text_size[0], text_size[1]
    text_x = int((width - width_text) / 2) + x
    text_y = int((height + height_text) / 2) + y
    cv2.putText(keyboard, text, (text_x, text_y), font_letter, font_scale, (255,0,0), font_th)

    
    
# letras(0,0, "Q")
# letras(0,0, "W")
# letras(0,0, "E")
frames = 0 
letter_index = 0

while True:
    keyboard[:] = (0,0,0)
    frames +=1
    
    
        # Letters
    if frames == 10:
        letter_index += 1
        frames = 0
    if letter_index == 15:
        letter_index = 0

    for i in range(15):
        if (i == letter_index):
            ligth = True
        else:
            ligth = False
        letras(i, keys_set_1[i], ligth)



    cv2.imshow("tablero", keyboard)
    key = cv2.waitKey(100)
    if key & 0xFF == ord('q'):
        break
        
        
if key & 0xFF == ord('q'):
    cv2.destroyAllWindows() 
```
